```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
def simulate_ar1(phi_0, phi_1, T, random_state, plot = True):
    y = np.zeros(T)
    if phi_1 == 1:
        y[0] = stats.norm.rvs(1)
    else:
        y[0] = phi_0 / (1-phi_1)
    epsilons = stats.norm.rvs(size = T, random_state = random_state)
    for t in range(1, T):
        y[t] = phi_0 + phi_1*y[t-1] + epsilons[t]
    if plot:
        fig = plt.figure(figsize = (8,4.5))
        plt.plot(y)
        plt.xlabel('Период')
        plt.ylabel('Значение')
        plt.title(r'Cимуляция процесса $Y_t = %.2f + %.2f Y_{t-1} + \varepsilon_t$' %(phi_0, phi_1))
        plt.hlines(y = y[0], xmin =0, xmax = T, color = 'r', linestyles = '--')
        plt.text(x = T*0.95, y = y[0]*1.1, s = r'$\frac{%.2f}{1-%.2f}$' %(phi_0, phi_1), color = 'r')
        plt.show()
    else:
        return y
```

```python
for phi_1 in [0, 0.5, 0.9, 0.99]:
    simulate_ar1(phi_0 = 1, phi_1 = phi_1, T = 500, random_state = 42)
```

```python
for phi_1 in [0.5, -0.5, -0.9, -0.99]:
    simulate_ar1(phi_0 = 1, phi_1 = phi_1, T = 500, random_state = 42)
```

```python
for phi_1 in [1, 1.05, 1.1, -1.1]:
    simulate_ar1(phi_0 = 1, phi_1 = phi_1, T = 100, random_state = 42)
```
