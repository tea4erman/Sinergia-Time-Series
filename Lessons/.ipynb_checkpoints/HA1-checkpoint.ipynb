```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

### Шаг 1. Загрузка данных

```python
import statsmodels.api as sm
item = 'prices'
package = 'fpp3'

df = sm.datasets.get_rdataset(item, package , cache=True).data
df = df.set_index('year')
df
```

### Шаг 2. Визуализация

```python
series = df['copper']
fig = plt.figure(figsize = (8,4.5))
plt.plot(series)
plt.xlabel('Год')
plt.ylabel('Долларов')
plt.title('Цена на медь')
plt.show()
```

```python
fig = plt.figure(figsize = (8,4.5))
plt.plot(series.diff())
plt.xlabel('Год')
plt.ylabel('Долларов')
plt.title('Первые разницы цены меди')
plt.show()
```
