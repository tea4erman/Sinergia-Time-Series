```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
acf = np.array([1, 0.9/1.81]+[0]*8)
```

```python
fig = plt.figure(figsize = (8,4.5))
plt.bar(x = range(len(acf)),height = acf)
plt.xlabel('Лаг (h)')
plt.ylabel('Автокорреляция')
plt.title(r'Коррелограмма для процесса $Y_t= \varepsilon_t+0.9\varepsilon_{t-1}$')
plt.show()
```
