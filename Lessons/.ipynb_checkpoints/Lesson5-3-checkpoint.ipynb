```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from statsmodels.tsa.stattools import acf
from statsmodels.tsa.stattools import pacf
```

```python
def plot_arma1_acf(phi_0, phi_1, theta_1, T, random_state):
    y = np.zeros(shape = (1000, T))
    y[:, 0] = phi_0 / (1-phi_1)
    epsilons = stats.norm.rvs(size = (1000, T), random_state = random_state)
    for t in range(1, T):
        y[:, t] = phi_0 + phi_1*y[:, t-1] + epsilons[:, t] + theta_1*epsilons[:, t-1]
        
    auto_cf = np.mean(np.apply_along_axis(lambda x: acf(x, nlags = 29), 1, y), axis = 0)
    auto_cf = np.where(np.abs(auto_cf)>0.02, auto_cf, 0)
    fig = plt.figure(figsize = (8,4.5))
    plt.bar(x = range(len(auto_cf)), height = auto_cf)
    sign_phi = np.where(phi_1 > 0, '+', '-')
    sign_theta = np.where(theta_1 > 0, '+', '-')
    plt.xlabel('Лаг (h)')
    plt.ylabel('Aвтокорреляция')
    plt.title(r'Коррелограмма АКФ для процесса $Y_t = %.2f %s %.2f Y_{t-1} + \varepsilon_t %s %.2f \varepsilon_{t-1} $' %(np.abs(phi_0), sign_phi, np.abs(phi_1), sign_theta, np.abs(theta_1)))
    plt.show()
```

```python
def plot_arma1_pacf(phi_0, phi_1, theta_1, T, random_state):
    y = np.zeros(shape = (1000, T))
    y[:, 0] = phi_0 / (1-phi_1)
    epsilons = stats.norm.rvs(size = (1000, T), random_state = random_state)
    for t in range(1, T):
        y[:, t] = phi_0 + phi_1*y[:, t-1] + epsilons[:, t] + theta_1*epsilons[:, t-1]
        
    partial_acf = np.mean(np.apply_along_axis(lambda x: pacf(x, nlags = 29), 1, y), axis = 0)
    partial_acf = np.where(np.abs(partial_acf)>0.02, partial_acf, 0)
    fig = plt.figure(figsize = (8,4.5))
    plt.bar(x = range(len(partial_acf)), height = partial_acf)
    sign_phi = np.where(phi_1 > 0, '+', '-')
    sign_theta = np.where(theta_1 > 0, '+', '-')
    plt.xlabel('Лаг (h)')
    plt.ylabel('Частичная автокорреляция')
    plt.title(r'Коррелограмма ЧАКФ для процесса $Y_t = %.2f %s %.2f Y_{t-1} + \varepsilon_t %s %.2f \varepsilon_{t-1} $' %(np.abs(phi_0), sign_phi, np.abs(phi_1), sign_theta, np.abs(theta_1)))
    plt.show()
```

```python
for phi_1 in [0.1, 0.5, 0.95]:
    for theta_1 in [0.1, 1, 2]:
        plot_arma1_acf(phi_0 = 1, phi_1 = phi_1, theta_1 = theta_1, T = 500, random_state = 42)
```

```python
for phi_1 in [0.1, 0.5, 0.95]:
    for theta_1 in [-0.1, -1, -2]:
        plot_arma1_acf(phi_0 = 1, phi_1 = phi_1, theta_1 = theta_1, T = 500, random_state = 42)
```
