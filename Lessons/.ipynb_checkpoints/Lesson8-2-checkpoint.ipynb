```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import statsmodels.api as sm
```

```python
def plot_acf_pacf(y, model_name):
    fig, ax = plt.subplots(1,2, figsize = (8, 4.5))
    plt.suptitle('Эмпирические АКФ и ЧАКФ для модели %s' %model_name)
    sm.graphics.tsa.plot_acf(y, ax = ax[0])
    ax[0].set_xlabel('Лаг')
    ax[0].set_ylabel('Автокорреляция')
    sm.graphics.tsa.plot_pacf(y, ax = ax[1])
    ax[1].set_xlabel('Лаг')
    ax[1].set_ylabel('Частичная автокорреляция')
    plt.tight_layout()
    plt.show()
```

```python
sales = pd.read_csv(r'https://raw.githubusercontent.com/tea4erman/Sinergia-Time-Series/refs/heads/main/Data/lesson1_vinyl_sales.csv', index_col = 0)
sales.index = pd.to_datetime(sales.index)
sales.index.freq = 'W-SUN'
sales.head()
```

```python
plot_acf_pacf(sales, 'продаж')
```

```python
evaluation = pd.DataFrame(columns = ['lagAR', 'lagMA', 'aic', 'bic'])
row = 0
for i in range(0,5):
    for j in range(0,5):
        results = sm.tsa.arima.ARIMA(sales, order=(i, 0, j)).fit()
        evaluation.loc[row, ['lagAR', 'lagMA', 'aic', 'bic']] = [i, j, results.aic, results.bic]
        row += 1
```

```python
evaluation.sort_values(by = 'aic').head()
```

```python
results = sm.tsa.arima.ARIMA(sales, order=(1, 0, 0)).fit()
print(results.summary())
```

```python
residuals = results.resid
sm.stats.acorr_ljungbox(residuals, lags = 10, boxpierce= True)
```
