```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import statsmodels.api as sm
```

```python
def simulate_ar1(phi_0, phi_1, T, random_state):
    y = np.zeros(T)
    if phi_1 == 1:
        y[0] = stats.norm.rvs(1)
    else:
        y[0] = phi_0 / (1-phi_1)
    epsilons = stats.norm.rvs(size = T, random_state = random_state)
    for t in range(1, T):
        y[t] = phi_0 + phi_1*y[t-1] + epsilons[t]
    return y
```

```python
def simulate_ar2(phi_0, phi_1, phi_2, T, random_state, plot = True):
    y = np.zeros(T)
    y[0] = phi_0 / (1-phi_1-phi_2)
    y[1] = phi_0 / (1-phi_1-phi_2)
    epsilons = stats.norm.rvs(size = T, random_state = random_state)
    for t in range(1, T):
        y[t] = phi_0 + phi_1*y[t-1] + phi_2*y[t-2] + epsilons[t]
    return y
```

```python
def simulate_ma1(mu, theta_1, T, random_state, plot = True):
    y = np.zeros(T)
    y[0] = mu
    epsilons = stats.norm.rvs(size = T, random_state = random_state)
    for t in range(1, T):
        y[t] = mu + epsilons[t] + theta_1*epsilons[t-1]
    return y
```

```python
def simulate_ma2(mu, theta_1, theta_2, T, random_state, plot = True):
    y = np.zeros(T)
    y[0] = mu
    y[1] = mu
    epsilons = stats.norm.rvs(size = T, random_state = random_state)
    for t in range(2, T):
        y[t] = mu + epsilons[t] + theta_1*epsilons[t-1] + theta_2*epsilons[t-2]
    return y
```

```python
def simulate_arma11(phi_0, phi_1, theta_1, T, random_state, plot = True):
    y = np.zeros(T)
    if phi_1 == 1:
        y[0] = stats.norm.rvs(1)
    else:
        y[0] = phi_0 / (1-phi_1)
    epsilons = stats.norm.rvs(size = T, random_state = random_state)
    for t in range(1, T):
        y[t] = phi_0 + phi_1*y[t-1] + epsilons[t] + theta_1*epsilons[t-1]
    return y
```

```python
def simulate_arma22(phi_0, phi_1, phi_2, theta_1, theta_2, T, random_state, plot = True):
    y = np.zeros(T)
    y[0] = phi_0 / (1-phi_1-phi_2)
    y[1] = phi_0 / (1-phi_1-phi_2)
    epsilons = stats.norm.rvs(size = T, random_state = random_state)
    for t in range(2, T):
        y[t] = phi_0 + phi_1*y[t-1] + phi_2*y[t-2] + epsilons[t] + theta_1*epsilons[t-1] + theta_2*epsilons[t-2]
    return y
```

```python
def plot_acf_pacf(y, model_name):
    fig, ax = plt.subplots(1,2, figsize = (8, 4.5))
    plt.suptitle('Эмпирические АКФ и ЧАКФ для модели %s' %model_name)
    sm.graphics.tsa.plot_acf(y, ax = ax[0])
    ax[0].set_xlabel('Лаг')
    ax[0].set_ylabel('Автокорреляция')
    sm.graphics.tsa.plot_pacf(y, ax = ax[1])
    ax[1].set_xlabel('Лаг')
    ax[1].set_ylabel('Частичная автокорреляция')
    plt.tight_layout()
    plt.show()
```

```python
def get_acf_pacf_table(y):
    y_acf = sm.tsa.stattools.acf(y,nlags = 10, alpha = 0.05)
    y_pacf = sm.tsa.stattools.pacf(y, nlags = 10, alpha = 0.05)
    acf_table = pd.DataFrame(data = np.round(y_acf[0],3))
    acf_table.columns = ['АКФ']
    acf_table['Нижняя граница ДИ для АКФ'] = np.round(y_acf[1][:,0],3)
    acf_table['Верхняя граница ДИ для АКФ'] = np.round(y_acf[1][:,1],3)
    acf_table['ЧАКФ'] = np.round(y_pacf[0],3)
    acf_table['Нижняя граница ДИ для ЧАКФ'] = np.round(y_pacf[1][:,0],3)
    acf_table['Верхняя граница ДИ для ЧАКФ'] = np.round(y_pacf[1][:,1],3)
    return acf_table
```

```python
phi_0 = 1
phi_1 = 0.5
T = 500
y = simulate_ar1(phi_0, phi_1, T, random_state = 42)
plot_acf_pacf(y, 'AR(1)')
get_acf_pacf_table(y)
```

```python
phi_0 = 1
phi_1 = 0.5
phi_2 = -0.4
T = 500
y = simulate_ar2(phi_0, phi_1, phi_2, T, random_state = 42)
plot_acf_pacf(y, 'AR(2)')
get_acf_pacf_table(y)
```

```python
mu = 1
theta_1 = 0.5
T = 500
y = simulate_ma1(mu, theta_1, T, random_state = 42)
plot_acf_pacf(y, 'MA(1)')
get_acf_pacf_table(y)
```

```python
mu = 1
theta_1 = 1
theta_2 = 0.4
T = 500
y = simulate_ma2(mu, theta_1, theta_2, T, random_state = 42)
plot_acf_pacf(y, 'MA(2)')
get_acf_pacf_table(y)
```

```python
phi_0 = 1
phi_1 = 0.5
theta_1 = 2
T = 500
y = simulate_arma11(phi_0, phi_1, theta_1, T, random_state = 42)
plot_acf_pacf(y, 'ARMA(1,1)')
get_acf_pacf_table(y)
```

```python
phi_0 = 1
phi_1 = -0.5
phi_2 = 0.4
theta_1 = 2
theta_2 = 0.5
T = 500
y = simulate_arma22(phi_0, phi_1, phi_2, theta_1, theta_2, T, random_state = 42)
plot_acf_pacf(y, 'ARMA(2,2)')
get_acf_pacf_table(y)
```
