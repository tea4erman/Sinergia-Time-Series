```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from statsmodels.tsa.stattools import acf
from statsmodels.tsa.stattools import pacf
```

### MA(2)

```python
def simulate_ma2(mu, theta_1, theta_2, T, random_state, plot = True):
    y = np.zeros(T)
    y[0] = mu
    y[1] = mu
    epsilons = stats.norm.rvs(size = T, random_state = random_state)
    for t in range(2, T):
        y[t] = mu + epsilons[t] + theta_1*epsilons[t-1] + theta_2*epsilons[t-2]
    if plot:
        fig = plt.figure(figsize = (8,4.5))
        plt.plot(y)
        plt.xlabel('Период')
        plt.ylabel('Значение')
        plt.title(r'Cимуляция процесса $Y_t = %.2f + \varepsilon_t + %.2f \varepsilon_{t-1} + %.2f \varepsilon_{t-2} $' %(mu, theta_1, theta_2))
        plt.hlines(y = y[0], xmin =0, xmax = T, color = 'r', linestyles = '--')
        #plt.text(x = T*0.95, y = y[0]*1.1, s = r'$\frac{%.2f}{1-%.2f}$' %(phi_0, phi_1), color = 'r')
        plt.show()
    else:
        return y
```

```python
for theta_1, theta_2 in [(2, 2), (2, 0.5), (0.5, 2), (0.5, 0.5)]:
    simulate_ma2(mu = 1, theta_1 = theta_1 , theta_2 = theta_2, T = 500, random_state = 42, plot = True)
```

```python
for theta_1, theta_2 in [(0.5, 0.5), (0.5, -0.5), (-0.5, 0.5), (-0.5, -0.5)]:
    simulate_ma2(mu = 1, theta_1 = theta_1 , theta_2 = theta_2, T = 500, random_state = 42, plot = True)
```

```python
def plot_ma2_acf(mu, theta_1, theta_2, T, random_state):
    y = np.zeros(shape = (1000, T))
    y[:, 0] = mu
    y[:, 1] = mu
    epsilons = stats.norm.rvs(size = (1000, T), random_state = random_state)
    for t in range(2, T):
        y[:, t] = mu + epsilons[:, t] + theta_1*epsilons[:, t-1] + theta_2*epsilons[:, t-2]
    auto_cf = np.mean(np.apply_along_axis(lambda x: acf(x, nlags = 29), 1, y), axis = 0)
    auto_cf = np.where(np.abs(auto_cf)>0.02, auto_cf, 0)
    fig = plt.figure(figsize = (8,4.5))
    plt.bar(x = range(len(auto_cf)), height = auto_cf)
    plt.xlabel('Лаг (h)')
    plt.ylabel('Aвтокорреляция')
    plt.title(r'Коррелограмма АКФ для процесса  $Y_t = %.2f + \varepsilon_t + %.2f \varepsilon_{t-1} + %.2f \varepsilon_{t-2} $' %(mu, theta_1, theta_2))
    plt.show()
```

```python
for theta_1, theta_2 in [(2, 2), (2, 0.5), (0.5, 2), (0.5, 0.5)]:
    plot_ma2_acf(mu = 1, theta_1 = theta_1 , theta_2 = theta_2, T = 500, random_state = 42)
```

```python
for theta_1, theta_2 in [(0.5, 0.5), (0.5, -0.5), (-0.5, 0.5), (-0.5, -0.5)]:
    plot_ma2_acf(mu = 1, theta_1 = theta_1 , theta_2 = theta_2, T = 500, random_state = 42)
```

```python
def plot_ma2_pacf(mu, theta_1, theta_2, T, random_state):
    y = np.zeros(shape = (1000, T))
    y[:, 0] = mu
    y[:, 1] = mu
    epsilons = stats.norm.rvs(size = (1000, T), random_state = random_state)
    for t in range(2, T):
        y[:, t] = mu + epsilons[:, t] + theta_1*epsilons[:, t-1] + theta_2*epsilons[:, t-2]
    partial_acf = np.mean(np.apply_along_axis(lambda x: pacf(x, nlags = 29), 1, y), axis = 0)
    partial_acf = np.where(np.abs(partial_acf)>0.02, partial_acf, 0)
    fig = plt.figure(figsize = (8,4.5))
    plt.bar(x = range(len(partial_acf)), height = partial_acf)
    plt.xlabel('Лаг (h)')
    plt.ylabel('Частичная автокорреляция')
    plt.title(r'Коррелограмма ЧАКФ для процесса  $Y_t = %.2f + \varepsilon_t + %.2f \varepsilon_{t-1} + %.2f \varepsilon_{t-2} $' %(mu, theta_1, theta_2))
    plt.show()
```

```python
for theta_1, theta_2 in [(2, 2), (2, 0.5), (0.5, 2), (0.5, 0.5)]:
    plot_ma2_pacf(mu = 1, theta_1 = theta_1 , theta_2 = theta_2, T = 500, random_state = 42)
```

```python
for theta_1, theta_2 in [(0.5, 0.5), (0.5, -0.5), (-0.5, 0.5), (-0.5, -0.5)]:
    plot_ma2_pacf(mu = 1, theta_1 = theta_1 , theta_2 = theta_2, T = 500, random_state = 42)
```
