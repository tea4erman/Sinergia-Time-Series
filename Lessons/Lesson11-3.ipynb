```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller
```

```python
def plot_acf_pacf(y, model_name):
    fig, ax = plt.subplots(1,2, figsize = (8, 4.5))
    plt.suptitle('Эмпирические АКФ и ЧАКФ для модели %s' %model_name)
    sm.graphics.tsa.plot_acf(y, ax = ax[0])
    ax[0].set_xlabel('Лаг')
    ax[0].set_ylabel('Автокорреляция')
    sm.graphics.tsa.plot_pacf(y, ax = ax[1])
    ax[1].set_xlabel('Лаг')
    ax[1].set_ylabel('Частичная автокорреляция')
    plt.tight_layout()
    plt.show()
    
def r_squared(y_true, y_pred):
    return 1 - (np.sum((y_true - y_pred)**2)/np.sum((y_true - np.mean(y_true))**2))

def mean_absolute_percentage_error(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

def root_mean_squared_error(y_true, y_pred):
    return np.sqrt(np.mean((y_true - y_pred)**2))

def plot_forecast(y_train, y_test, y_forecast, trim = 300):
    T = len(y_train)
    horizon = len(y_test)
    fig = plt.figure(figsize = (8,4.5))
    plt.plot(range(trim, T), y_train[trim:], label = 'train')
    plt.xlabel('Период')
    plt.ylabel('Значение')
    plt.title(r'Тренинговая, тестовая выборки и прогноз')
    plt.plot(range(T, T + horizon), y_test, label = 'test')
    plt.plot(range(T, T + horizon), y_forecast, label = 'forecast')
    plt.legend()
    plt.show()
```

```python
y = usdrub.loc['2002':'2012'].diff().dropna().values.ravel()
```

```python
periods = []
QLRstat = []
T = len(y)
tau0 = int(np.ceil(0.15*T))
tau1 = int(np.floor(0.85*T))

for t in range(tau0,tau1):
    d1 = np.zeros(T)
    d1[t:] = 1
    Y = y[1:]
    X = np.vstack((y[:-1], d1[1:], y[:-1]*d1[1:])).T
    X = sm.add_constant(X)
    model = sm.OLS(Y,X)
    results = model.fit()
    f_test = results.f_test([0,0,1,1])
    fvalue = f_test.fvalue
    periods.append(t)
    QLRstat.append(fvalue)

QLR_statistics = pd.DataFrame({'t': periods, 'QLRstat': QLRstat})
```

```python
subset = usdrub.loc['2009-02-19':'2013']
```

```python
fig = plt.figure(figsize = (8,4.5))
plt.plot(subset)
plt.xlabel('День')
plt.ylabel('Курс RUR / USD')
plt.title('Динамика курса доллара США c 2009-02-19 до 2013-12-31')
plt.show()
```

```python
adftest = adfuller(subset.dropna(), maxlag = 10, regression = 'c')

dfoutput = pd.Series(
        adftest[0:4],
        index=[
            "Test Statistic",
            "p-value",
            "#Lags Used",
            "Number of Observations Used",
        ]
)

dfoutput
```

```python
plot_acf_pacf(subset.dropna(), 'курса доллара')
```

```python
evaluation = pd.DataFrame(columns = ['lagAR', 'lagMA', 'aic', 'bic'])
row = 0
for i in range(0,5):
    for j in range(0,5):
        results = sm.tsa.arima.ARIMA(subset.dropna(), order=(i, 0, j)).fit()
        evaluation.loc[row, ['lagAR', 'lagMA', 'aic', 'bic']] = [i, j, results.aic, results.bic]
        row += 1
        
evaluation.sort_values(by = 'bic').head()
```

```python
results = sm.tsa.arima.ARIMA(subset.dropna(), order=(1, 0, 0)).fit()
print(results.summary())
```

```python
sm.stats.acorr_ljungbox(results.resid, lags = 10, boxpierce= True)
```

```python
y_train = subset.iloc[:-10].values
y_test = subset.iloc[-10:].values
results = sm.tsa.arima.ARIMA(y_train, order=(1, 0, 0)).fit()
y_fcst = results.forecast(steps = 10)
plot_forecast(y_train, y_test, y_fcst, trim = 1100)
print('R-squared:', r_squared(y_test, y_fcst))
print('RMSE:', root_mean_squared_error(y_test, y_fcst))
print('MAPE:', mean_absolute_percentage_error(y_test, y_fcst))
```
